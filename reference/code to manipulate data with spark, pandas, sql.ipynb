```python
import findspark
findspark.init()
import pyspark
import random
sc = pyspark.SparkContext(appName="Pi")
import pandas as pd
```

```python
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("Python Spark SQL basic example") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
```

```python
flagging = spark.read.csv("data/data.csv", header =False)
```

```python
flagging
```

```python
from pyspark.sql.functions import *

flagging = flagging.select(col("_c0").alias("Prod_ID"), col("_c1").alias("Date"), col("_c2").alias("IPU_Ver"),col("_c3").alias("Val"), col("_c4").alias("Counter") )
```

```python
flagging.write.saveAsTable("Flagging")
```

```python
flagging.show()
```

```python
sqlDF = spark.sql("SELECT * FROM Flagging where IPU_Ver ='00-19 (Build 9)'")
flagging_df =sqlDF.toPandas()
```
